```python
import matplotlib as plt
import pandas as pd
df=pd.read_csv("clevelanda (1).csv")
df
```

```python
import matplotlib.pyplot as plt
plt.bar(df['age'], df['chol'])
plt.xlabel('age')
plt.ylabel('chol')
plt.show()
```

```python
plt.plot(df['slope'],df["cp"])
plt.title("line plot between slope and cp")
plt.xlabel("age")
plt.ylabel("slope")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,7,8,10]
plt.plot(x,y,color='green')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,7,8,10]
plt.plot(x,y,linestyle="dashed")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,7,8,10]
plt.plot(x,y,marker="*")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,7,8,10]
plt.plot(x,y,linewidth="0.5",color="green",marker="*")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[10000,40000,10000,20000,50000,5000]
plt.plot(x,y,linestyle="dashed",color="green",marker="o")
plt.title("saving amounts by month")
plt.xlabel("months")
plt.ylabel("saving amounts")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[10000,40000,10000,20000,50000,5000]
plt.bar(x,y,color="lightgreen")
plt.title("saving amounts by month")
plt.xlabel("months")
plt.ylabel("saving amounts")
plt.show()
```

```python
ages=[20,30,40,10,30,30,40,10,60,60]
b=[10,40,70,100]
#bins are the intervals for grouping the data
plt.hist(ages,bins=b,color="orange",edgecolor="green")
plt.xlabel("ages")
plt.ylabel("b")
plt.grid(True)
plt.show()
```

```python
x=['python','java','c','javascript']
y=[25,40,5,30]
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.title("pie chart")
plt.legend(loc="upper left")
plt.show()
```
